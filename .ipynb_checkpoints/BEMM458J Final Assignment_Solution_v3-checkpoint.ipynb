## Task 1. Which Twitter users are the most popular ? (10%)
### For each Task please: (1) develop the code required and (2) provide a brief discussion and interpretation of the results

### Tip: consider retweet and favorite counts as proxies for popularity

<h2><span style="color: #008000;"><strong>1a. Sorting the sum of retweet and favourite counts</strong></span></h2>

```python
#Group by the user name the sum of the retweet and favourite counts colums
task1=ConversationsLean.groupby(['tweet.user.screen_name']).sum()[['tweet.retweet_count','tweet.user.favourites_count']]
```

```python
#Sort the value by retweets
task1.sort_values(by='tweet.retweet_count',ascending=False).head()
```

```python
#Sort the value by favourites
task1.sort_values(by='tweet.user.favourites_count',ascending=False).head()
```

<h3><span style="color: #008000;">The tables show the sorted sum of the retweet and  favourite counts per user. It is listed the three most popular users for both options are: PlasticPollutes, Greenpeace and 5gyres. However, it is not so clear which user is more popular in the fourth place, so we suggest the method shown in item 1b.</span></h3>
<br>


<h2><span style="color: #008000;"><strong>1b. Calculating the weighted linear combination of retweets and favourite counts  (50% -50%)</strong></span></h2>

```python
# Sum all the values for the coulumns retweet and favorite
sumfav=task1[['tweet.user.favourites_count']].sum()
sumret=task1[['tweet.retweet_count']].sum()
# Normalise each column dividing them by its sum
retweet_norm=task1['tweet.retweet_count'].apply(lambda x: x/sumret)
fav_norm=task1['tweet.user.favourites_count'].apply(lambda x: x/sumfav)
# Merge the columns into a new dataframe
df_merge_col = pd.merge(fav_norm, retweet_norm, on='tweet.user.screen_name')
#Calculate the weighted linear combination  of retweets and favourite counts (50%-50%)
df_merge_col['weighted_col']=df_merge_col['tweet.user.favourites_count']*.5+df_merge_col['tweet.retweet_count']*.5
#Sort the values
df_merge_col.sort_values(by='weighted_col',ascending=False).head(10)['weighted_col']
```

<h3><span style="color: #008000;">Calculating the weighted linear combination of retweets and favourite counts, we can define the most popular users in the whole table by giving a specific weight to retweets and favourite counts. In this case, we gave equal weight to each element, but the marketing department staff can add some specific weights. This calculation is achieved by normalizing both columns, forcing them to sum one; then calculte the sum of the columns multiplied by the chosen weight. This value can be seen in the weighted_col column.</span></h3>

## Task 3. Do emotions play a role in the number of retweets and favorites achieved by tweets ? (20 %)
### For each Task please: (1) develop the code required and (2) provide a brief discussion and interpretation of the results

### Tip:  correlations between variables
### Tip:  visualizations relating variables
### Tip:  optionally consider basic regression models to determine the impact of some variables on others (e.g. impact of love on favorite, impact of fear on retweet)

```python
task3=ConversationsLean.groupby(['tweet.retweet_count','tweet.user.favourites_count']).mean()[['anger','fear','joy','love','sadness','trust','identity_hate', 'insult', 'obscene', 'severe_toxic', 'threat', 'toxic']].reset_index()
task3.columns = ['retweet', 'favourites','anger','fear','joy','love','sadness','trust','id_hate', 'insult', 'obscene', 'toxic', 'threat', 'toxic']
task3=task3.astype(float)
task3
```

<h2><span style="color: #008000;"><strong>3a. Create a Dataframe with the emotions</strong></span></h2>

```python
# Groupby Data
task3_retweet=ConversationsLean.groupby(['tweet.retweet_count']).mean()[['anger','fear','joy','love','sadness','trust','identity_hate', 'insult', 'obscene', 'severe_toxic', 'threat', 'toxic']].reset_index()
# set the data type
task3_retweet = task3_retweet.astype(float)
#Change column names
task3_retweet.columns = ['retweet', 'anger','fear','joy','love','sadness','trust','id_hate', 'insult', 'obscene', 'sevtoxic', 'threat', 'toxic']
#Set retweet as index
task3_retweet = task3_retweet.set_index('retweet')
task3_retweet
```

<h2><span style="color: #008000;"><strong>3b. Calculate the multicollinearity in the data</strong></span></h2>

```python
# calculate the correlation matrix  between variables
corr=task3_retweet.corr()
display(corr)
#Plot correlation
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

<h3><span style="color: #008000;">We can observe the variables that are highly correlated, we will drop that variables.</span></h3>
<br>

<h2><span style="color: #008000;"><strong>3c. Calculate variance inflation factort to make sure that the eliminated variables do not contribute to the standar error in our model.</strong></span></h2>

```python
#Calculate the Variance Inflation factor
task3_retweet_before = task3_retweet
X1 = sm.tools.add_constant(task3_retweet_before)
task3_before = pd.Series([variance_inflation_factor(X1.values, i) for i in range(X1.shape[1])], index=X1.columns)
display(task3_before)
```

```python
# Eliminate the variables that are high correlated
task3_retweet_after = task3_retweet.drop(['toxic','joy', 'id_hate','insult','sevtoxic','obscene'], axis = 1)
# Caculate the Variance Inflation factor again
X2 = sm.tools.add_constant(task3_retweet_after)
task3_after = pd.Series([variance_inflation_factor(X2.values, i) for i in range(X2.shape[1])], index=X2.columns)
display(task3_after)
```

```python
task3_retweet_after.head()
# calculate the correlation matrix with the selected variables
corr=task3_retweet_after.corr()
display(corr)
#Plot correlation
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, cmap='RdBu')
```

```python
# Plot the relation of the variables
pd.plotting.scatter_matrix(task3_retweet_after, alpha = 1, figsize = (14,9))
plt.show()
```

<h3><span style="color: #008000;">Now we have the variables that we will use in our regression model.</span></h3>

<h2><span style="color: #008000;"><strong>3d. Regression Model for retweets</strong></span></h2>

```python
retweet_osl = sm.ols( formula="retweet ~ anger+fear+love+sadness+trust+threat",data=task3).fit()
print(retweet_osl.summary())
```
